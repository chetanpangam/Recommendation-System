```python
#floats.extend([float(x) for x in s.split(' ')])
lines = loadtxt("test5.txt", delimiter=" ", unpack=False)
#for s in lines:
print(lines[0][0])
```

```python
my_dict = defaultdict(list)
rows=len(lines)
for i in range(rows):
    arr=lines[i][1],lines[i][2]
    my_dict[lines[i][0]].append(arr)
```

```python
#print(my_dict.get(201))
ratingmat=[[0 for i in range(201)]for i in range(5)]
active_user=[]
for i in range(5):
    #print((my_dict.get(201))[i][0])
    active_user.append((my_dict.get(201))[i][1])  #contains the ranking given by the active user
    movie=((my_dict.get(201))[i][0])   #gets the movie which are predicted by the active user
    print(movie)
    print((my_dict.get(201))[i][1])
    #print(result[199][237])
    for j in range(200):
        ratingmat[i][j]=result[j][(movie-1).astype(int)]
orig_mat=(numpy.transpose(ratingmat))   #matrix of ranking of all users from training data for the movie rated by the test user
print()
print(active_user)
```

```python
import numpy  as np
vict=[]   #row vector of all the weights
for i in range(200):
    vict.append(cosine_similarity(np.array(active_user),np.array(orig_mat[i])))
#print(vict[0])
#print(len(vict))
```

```python
#Calculating the weight vector
for i in range(len(vict)):
    nonzeroind = np.nonzero(orig_mat[i])[0] #to get non zero index of the orig_matrix
    sum=0
    for index in nonzeroind:
        sum=sum+(active_user[index]*active_user[index])
    sum=math.sqrt(sum)            
    vict[i]=(vict[i]*np.linalg.norm(active_user - np.zeros(5)))/sum    #to neglect those terms which are not present in training user matrix
    #print(len(vict))
```

```python
#Prediction Block

n=0
#print (result[0][110])
#print(result[0][(my_dict.get(201))[6][0]])
for i in range(5,len(my_dict.get(201))):
    total_weight=0
    total=0
    for j in range(200):
        #print(vict[j]=="nan")
        if(((result[j][(my_dict.get(201))[i][0]-1])!=0) and ((~numpy.isnan(vict[j])))): # rating given by jth user to ith movie
            #print((my_dict.get(201))[i][0])
            #print(result[j][(my_dict.get(201))[i][0]-1])
            #print (j)
            total+=vict[j]*result[j][(my_dict.get(201))[i][0]-1]
            total_weight+=vict[j]
    predicted_value=(numpy.round(total/total_weight)).astype(int)
    print(predicted_value)
```
